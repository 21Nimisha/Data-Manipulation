```python
# find the  highest rated movie in theâ€œQuestâ€ story type.
```

```python
import numpy as np
import pandas as pd

dataset=pd.read_csv('HollywoodMovies.csv')
selected_data=dataset.loc[:,['Movie','Story','RottenTomatoes']]
df=pd.DataFrame(selected_data)
df_movie_filter = df[(df.Story == 'Quest') & (df.Story.notnull())]
output = df_movie_filter[df_movie_filter.RottenTomatoes == max(df_movie_filter.RottenTomatoes)]
print(output)
```

```python
#Is there any correspondence between the criticsâ€™ evaluation of a movie and itsacceptance by the public? Find out, by plotting the net profitability of a movie against the ratings it receives on Rotten Tomatoes
```

```python
import matplotlib.pyplot as plt, pandas

frame = pandas.read_csv('HollywoodMovies.csv', delimiter=',')

selected = frame.loc[:, ["Profitability", "RottenTomatoes"]]

plt.scatter(selected["Profitability"], selected["RottenTomatoes"])
plt.xlim(0, 200)
plt.show()
```
