```python
# Write a program to find the list of all countries with a size greater than 2000(*1000) square km?
```

```python
# Countries with larger area

selected_data = data.loc[: , ['Country', 'LandArea']]  #.loc selcet only specific columns

for i in selected_data.itertuples():
    if i[2]> 2000:
        print(i)
        
# output contains all countries with area > 2000 (*1000)km        
```

```python
# Is there a correlation between the GDP per capita of countries and their birth rates?
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (50,50))
data = pd.read_csv("AllCountries.csv")
selected_data = data.loc[: , ['Country', 'GDP', 'BirthRate']]
x = np.array(selected_data['GDP'])
y = np.array(selected_data['BirthRate'])

plt.scatter(x,y)

plt.xlim(0,20000)
plt.show()


# this plot shows that the birth rate of the countries declines with the increase in GDP.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (50,50))
data = pd.read_csv('AllCountries.csv')
selected_data = data.loc[:, ['Country', 'GDP', 'BirthRate']]
sortdata = selected_data.sort_index(by = 'GDP', ascending = False)
selected_sortdata = sortdata.iloc[:10] #.iloc select the 10 first row of the sorted dataset
print(selected_sortdata)
```

```python
plt.pie(selected_sortdata['GDP'], labels = selected_sortdata['Country'])
plt.show()
```
